```python
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
model.fit(xs, ys, epochs=500)
```

```python
mnist = tf.keras.datasets.fashion_mnist
(training_images, training_labels), (test_images, test_labels) = mnist.load_data()
```

```python
np.set_printoptions(linewidth=200)
import matplotlib.pyplot as plt
plt.imshow(training_images[0])
print(training_images.shape)
print(training_labels[0])
print(training_images[0])
```

```python
training_images = training_images / 255.0
test_images = test_images / 255.0
```

```python
model = tf.keras.models.Sequential([tf.keras.layers.Flatten(),
                                    tf.keras.layers.Dense(128, activation=tf.nn.relu),
                                    tf.keras.layers.Dense(10, activation=tf.nn.softmax)])
```

```python
class myCallback(tf.keras.callbacks.Callback):  # 'callback' ❌ → 'callbacks' ✅
    def on_epoch_end(self, epoch, logs={}):  # 'om_epoch_end' ❌ → 'on_epoch_end' ✅
        if logs.get('accuracy') > 0.95:
            print("\nReached 95% accuracy, stopping training!")
            self.model.stop_training = True
```

```python
model.compile(optimizer = tf.optimizers.Adam(),
              loss = 'sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(training_images, training_labels, epochs=20, callbacks=[myCallback()])  # ✅ ')' बंद कर दी
```

```python
model.evaluate(test_images, test_labels)
```

```python
mmnist = tf.keras.datasets.mnist
(X_train, y_train), (X_test, y_test) = mmnist.load_data()
```

```python
np.set_printoptions(linewidth=200)
plt.imshow(X_train[0])
print(X_train.shape)
print(X_test.shape)
print(y_train[0])
print(X_train[0])
```

```python
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
model = tf.keras.models.Sequential([tf.keras.layers.Flatten(input_shape = (28,28)),
                                    tf.keras.layers.Dense(525, activation=tf.nn.relu),
                                    tf.keras.layers.Dense(10, activation=tf.nn.softmax)])

```

```python
class myCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs={}):
        if logs.get('accuracy', 0) > 0.99:
            print("\nReached 99% accuracy, stopping training!")
            self.model.stop_training = True
model.fit(X_train, y_train, epochs=20, callbacks=[myCallback()])
```

```python
model.evaluate(X_test, y_test)
```
